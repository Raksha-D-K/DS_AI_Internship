```python
1. Supervised Learning
Supervised learning is type of Machine Learning where the dataset contains input data + correct output
The model learns to predict answers.
2.UnSupervised Learning
Machine learning where data has no labels.
The model finds hidden patterns or clusters.
3.Reinforcement Learning
Machine learning where an agent learns using rewards and penalties through interaction with an environment.
```

```python
| ML Type       | Data Type    | Example 1             | Example 2               |
| ------------- | ------------ | --------------------- | ----------------------- |
| Supervised    | Labeled      | Spam detection        | House price prediction  |
| Unsupervised  | Unlabeled    | Customer segmentation | Recommendation grouping |
| Reinforcement | Reward-based | Self-driving cars     | Game playing AI         |
```

```python
Choosing the wrong ML type leads to failure because:
Using unsupervised learning when labels exist  loss of predictive power
Using supervised learning without labels, the model cannot learn
Using RL for prediction tasks results in unnecessary complexity and slow training
```
